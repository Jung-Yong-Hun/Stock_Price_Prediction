```python
#Import the pandas library
import pandas as pd
```

```python
#Load the data
from google.colab import files
files.upload()
```

```python
#Read in the data
df = pd.read_csv('AAPL.csv')
#Show the data
df
```

```python
#Select the date and price
df = df[['Date', 'Close']]
#Rename the features:
df = df.rename(columns={'Date':'ds', 'Close':'y'})
#Show the data
df
```

```python
#Get the last 20 rows of data and store them into a new variable
last = df[len(df)-20:]
last
```

```python
#Get all of the rows in the data set except for the last 20 rows of data
df = df[:-20]
df
```

```python
#Getting the Facebook prophet library
from fbprophet import Prophet
#Create the Prophet object (model)
fbp = Prophet(daily_seasonality = True)
#Fit or train the model
fbp.fit(df)
future = fbp.make_future_dataframe(periods=365)
forecast = fbp.predict(future)
```

```python
#Import thet facebooks prophet plotting libraries
from fbprophet.plot import plot_plotly
#plot the data
plot_plotly(fbp, forecast)
```

```python
#Show the models prediction for 10-15-2021
forecast[forecast.ds == '2021-10-15']['yhat']
```

```python
#Show the actual value of the stock for 10-15-2021
last[last.ds == '2021-10-15']['y']
```
